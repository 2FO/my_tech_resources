```python
assets = pd.DataFrame(series)
assets
```

```python
asset_list = list(assets.columns.values)
assets.tail()
```

```python
logged_assets = np.log(assets).dropna().diff().dropna()
```

```python
logged_assets
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=3)
pca.fit(logged_assets)
print(pca.explained_variance_ratio_)
```

```python
print(pca.components_)
```

```python
transformed_logged_assets = pca.transform(logged_assets)
transformed_logged_assets
```

```python
plt.scatter(transformed_logged_assets[:,0],transformed_logged_assets[:,1])
pylab.title('Principal components of Propane, Gasoline, Brent & W Texas Crude')
pylab.xlabel('1st principal component')
pylab.ylabel('2nd principal component')
```

```python
var_model = VAR(logged_assets)
var_results = var_model.fit(2)
var_results.summary()
```

```python
var_results.plot()
```

```python
var_results.plot_acorr()
```

```python
var_model.select_order(10)
```

```python
var_results = var_model.fit(maxlags=10, ic='aic')
```

```python
lag_order = var_results.k_ar
var_results.forecast(logged_assets.values[-lag_order:], 5)
```

```python
var_results.plot_forecast(2)
```

```python
irf = var_results.irf(10)
```
