```python
# based on example at http://statsmodels.sourceforge.net/devel/vector_ar.html

import statsmodels.api as sm
from statsmodels.tsa.api import VAR
import pandas as pd
import numpy as np
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.figsize'] = 18, 8

macro = sm.datasets.macrodata.load_pandas().data

dates = macro[['year', 'quarter']].astype(int).astype(str)
quarterly = dates["year"] + "Q" + dates["quarter"]
from statsmodels.tsa.base.datetools import dates_from_str
quarterly = dates_from_str(quarterly)

macro = macro[['realgdp','realcons','realinv']]
macro.index = pd.DatetimeIndex(quarterly)
logged_macro = np.log(macro).diff().dropna()

model = VAR(logged_macro) #sm.tsa.vector_ar.var_model.VAR(data)
```

```python
results = model.fit(2)
results.summary()
```

```python
results.plot()
```

```python
results.plot_acorr()
```

```python
model.select_order(15)
```

```python
results = model.fit(maxlags=15, ic='aic')
```

```python
lag_order = results.k_ar
results.forecast(logged_macro.values[-lag_order:], 5)
```

```python
results.plot_forecast(2)
```

```python
irf = results.irf(10)
```

```python
irf.plot(orth=False)
```

```python
irf.plot(impulse='realcons')
```

```python
irf.plot_cum_effects(orth=False)
```

```python
forecast_error_variance_decomp = results.fevd(5)
```

```python
forecast_error_variance_decomp.summary()
```

```python
results.fevd(20).plot()
```

```python
results.test_causality('realgdp', ['realinv', 'realcons'], kind='f')
```

```python
results.test_causality('realinv', ['realgdp', 'realcons'], kind='f')
```

```python
results.test_causality('realcons', ['realinv', 'realgdp'], kind='f')
```
