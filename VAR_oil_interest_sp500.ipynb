```python
from hedgeye_stock_analysis import *

import statsmodels.api as sm
from statsmodels.tsa.api import VAR
import pandas as pd
import numpy as np
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.figsize'] = 18, 8
```

```python
start = datetime.date(2000, 1, 1)
end = datetime.date.today() - datetime.timedelta(days = 1)

sp500 = get_stock("%5EGSPC", "data/sp500.csv")
sp500_close = sp500['Close']

#aapl = get_stock("AAPL", "data/aapl.csv")
#aapl_close = aapl['Close']

aal = get_stock("AAL", "data/aal.csv")
aal_close = aal['Close']

xom = get_stock("XOM", "data/xom.csv")
xom_close = xom['Close']

#usdollar = get_stock("%5EUSDOLLAR", "data/usdollar.csv")
#usdollar_close = usdollar['Close']

tenyrint = get_stock("%5ETNX", "data/tenyrint.csv")
tenyrint_close = tenyrint['Close']

vix = get_stock("%5EVIX", "data/vix.csv")
vix_close = vix['Close']

gold = get_stock("GLD", "data/gld.csv")
gold_close = gold['Close']

oil = get_stock("OIL", "data/oil.csv")
oil_close = oil['Close']

series = { 'sp500': sp500_close,
           'tenyrint': tenyrint_close,
           'oil': oil_close,
           #'gold': gold_close
           #'aal': aal_close
           'xom': xom_close
           #'usdollar': usdollar_close,
           #'vix': vix_close
         }

assets = pd.DataFrame(series)
asset_list = list(assets.columns.values)
assets.tail()
```

```python
logged_assets = np.log(assets).diff().dropna()
model = VAR(logged_assets)
```

```python
logged_assets
```

```python
results = model.fit(2)
results.summary()
```

```python
results.plot()
```

```python
results.plot_acorr()
```

```python
model.select_order(15)
```

```python
results = model.fit(maxlags=15, ic='aic')
```

```python
lag_order = results.k_ar
results.forecast(logged_assets.values[-lag_order:], 5)
```
